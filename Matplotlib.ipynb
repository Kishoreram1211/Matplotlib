Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#plotting x and y
plt.plot(x,y)
plt.show()
```

```python
#plotting x and z
plt.plot(x,z)
plt.show()
```

```python
#adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.title('Sine wave')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.show()
```

```python
#plotting a parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.plot(x,y,'b.')
plt.show()
```

```python
plt.plot(x,y,'gx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'b-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
